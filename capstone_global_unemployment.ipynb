# Finding the average in terms of sex per year for the whole dataset

```python
%%sql

SELECT distinct
sex,
age_group, 
age_categories,
AVG(year_2014) OVER(PARTITION BY sex, age_group ORDER BY age_group ) AS Avg_per_group2014,
AVG(year_2015) OVER(PARTITION BY sex, age_group ORDER BY age_group ) AS Avg_per_group2015,
AVG(year_2016) OVER(PARTITION BY sex, age_group ORDER BY age_group ) AS Avg_per_group2016,
AVG(year_2017) OVER(PARTITION BY sex, age_group ORDER BY age_group) AS Avg_per_group2017,
AVG(year_2018) OVER(PARTITION BY sex, age_group ORDER BY age_group) AS Avg_per_group2018,
AVG(year_2019) OVER(PARTITION BY sex, age_group ORDER BY age_group) AS Avg_per_group2019,
AVG(year_2020) OVER(PARTITION BY sex, age_group ORDER BY age_group) AS Avg_per_group2020,
AVG(year_2021) OVER(PARTITION BY sex, age_group ORDER BY age_group) AS Avg_per_group2021,
AVG(year_2022) OVER(PARTITION BY sex, age_group ORDER BY age_group) AS Avg_per_group2022,
AVG(year_2023) OVER(PARTITION BY sex, age_group ORDER BY age_group ) AS Avg_per_group2023,
AVG(year_2024) OVER(PARTITION BY sex, age_group ORDER BY age_group ) AS Avg_per_group2024

FROM global_unemployment
;
```

# Top 10 Countries experiencing the lowest rate of unemployment

```python
%%sql
WITH group_avg AS (
    SELECT
        country_name,
        sex,
        age_group,
        SUM(year_2014 + year_2015 + year_2016 + year_2017 + year_2018 +
            year_2019 + year_2020 + year_2021 + year_2022 + year_2023 + year_2024) / 11 AS Avg_per_group
    FROM global_unemployment
    GROUP BY country_name, sex, age_group
),
country_avg AS (
    SELECT
        country_name,
        AVG(Avg_per_group) AS avg_unemployment
    FROM group_avg
    GROUP BY country_name
)
SELECT *
FROM country_avg
ORDER BY avg_unemployment ASC
LIMIT 10;

```

```python
%%sql
WITH group_avg AS (
    SELECT
        country_name,
        sex,
        age_group,
        SUM(year_2014 + year_2015 + year_2016 + year_2017 + year_2018 +
            year_2019 + year_2020 + year_2021 + year_2022 + year_2023 + year_2024) / 11 AS Avg_per_group
    FROM global_unemployment
    GROUP BY country_name, sex, age_group
),
country_avg AS (
    SELECT
        country_name,
        AVG(Avg_per_group) AS avg_unemployment
    FROM group_avg
    GROUP BY country_name
)
SELECT *
FROM country_avg
ORDER BY avg_unemployment DESC
LIMIT 10;
```

# Ranking the unemployment levels for countries in descending order 

```python
%%sql
WITH group_avg AS (
    SELECT
        country_name,
        sex,
        age_group,
        SUM(year_2014 + year_2015 + year_2016 + year_2017 + year_2018 +
            year_2019 + year_2020 + year_2021 + year_2022 + year_2023 + year_2024) / 11 AS Avg_per_group
    FROM global_unemployment
    GROUP BY country_name, sex, age_group
),
country_avg AS (
    SELECT
        country_name,
        AVG(Avg_per_group) AS avg_unemployment
    FROM group_avg
    GROUP BY country_name
),
ranked_countries AS (
    SELECT
        country_name,
        avg_unemployment,
        RANK() OVER (ORDER BY avg_unemployment DESC) AS unemployment_rank
    FROM country_avg
)
SELECT *
FROM ranked_countries
WHERE unemployment_rank <=10;
```
